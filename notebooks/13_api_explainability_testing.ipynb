# Experiment 16: SHAP Driver Extraction Testing

This experiment validates the SHAP-based driver extraction logic that will be integrated into the prediction API.

The goal is to identify the top features contributing to the failure risk prediction for an individual machine.

```python
# ==================================================
# Imports
# ==================================================

from pathlib import Path
import sys

project_root = Path.cwd().parent

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from sklearn.model_selection import train_test_split

from src.data.load_data import (
    load_ai4i_dataset
)

from src.features.feature_engineering import (
    create_engineered_features
)

from src.features.preprocessing import (
    prepare_modeling_dataset
)

from src.models.champion_model import (
    build_champion_xgboost_model
)

from src.explainability.shap_analysis import (
    get_top_drivers
)


# ==================================================
# Load Data
# ==================================================

df = load_ai4i_dataset()

df = create_engineered_features(df)

df = prepare_modeling_dataset(df)

X = df.drop(
    columns=["Machine failure"]
)

y = df["Machine failure"]


# ==================================================
# Train Test Split
# ==================================================

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)


# ==================================================
# Load Champion Model
# ==================================================

model = build_champion_xgboost_model()

model.fit(
    X_train,
    y_train
)


# ==================================================
# Test Single Machine
# ==================================================

sample = X_test.iloc[[0]]

print("Machine Features:")
display(sample)


# ==================================================
# SHAP Driver Extraction
# ==================================================

drivers = get_top_drivers(

    model,

    sample,

    top_n=5

)

print("\nTop Drivers:\n")

display(drivers)
```

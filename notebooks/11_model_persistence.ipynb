# Experiment 14: Model Persistence

This experiment saves the champion model to disk and verifies that it can be loaded for future inference without retraining.

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.data.load_data import (
    load_ai4i_dataset
)

from src.features.feature_engineering import (
    create_engineered_features
)

from src.features.preprocessing import (
    prepare_modeling_dataset
)

from src.models.champion_model import (
    build_champion_xgboost_model
)

from src.models.model_registry import (
    save_model,
    load_model
)

from sklearn.model_selection import (
    train_test_split
)

# Pipeline

df = load_ai4i_dataset()

df = create_engineered_features(df)

df = prepare_modeling_dataset(df)

X = df.drop(
    columns=["Machine failure"]
)

y = df["Machine failure"]

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y
)

# Train

model = build_champion_xgboost_model()

model.fit(
    X_train,
    y_train
)

# Save

save_model(
    model,
    "champion_model"
)

# Load

loaded_model = load_model(
    "champion_model"
)

# Verify

probability = loaded_model.predict_proba(
    X_test.iloc[[0]]
)[0][1]

print(
    "Loaded model probability:",
    probability
)
```

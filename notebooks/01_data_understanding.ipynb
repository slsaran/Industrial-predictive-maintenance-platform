# Numerical Feature Summary by Failure Status

```python
numeric_features = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]"
]

df.groupby("Machine failure")[numeric_features].mean().T
```

```python
df.groupby("Machine failure")[numeric_features].median().T
```

Potential Engineered Features

1. Temperature Difference
   = Process Temperature - Air Temperature

2. Power Proxy
   = Torque × Rotational Speed

3. Mechanical Stress Index
   = Torque / Rotational Speed

4. Tool Wear Severity
   = Normalized Tool Wear

5. Thermal Stress Index
   = Temperature Difference × Torque

Feature Importance Hypothesis (Based on EDA)

The following features appear most likely to influence machine failures:

1. Torque
2. Tool Wear
3. Rotational Speed
4. Air Temperature
5. Process Temperature

This hypothesis will later be validated using:
- XGBoost Feature Importance
- SHAP Analysis

# Experiment 15: MLflow Experiment Tracking

This experiment tracks model parameters, metrics and artifacts using MLflow to support reproducibility and experiment management.

```python
import mlflow

mlflow.set_experiment(
    "predictive-maintenance"
)

with mlflow.start_run():

    mlflow.log_param(
        "max_depth",
        4
    )

    mlflow.log_param(
        "n_estimators",
        100
    )

    mlflow.log_metric(
        "f1_score",
        0.8615
    )

print(
    "Run logged successfully"
)
```
